## Data Analysis

### 1. Univariate Analysis - Analysing single column

```python
df.head(1)
```

Visualizing the distribution of price

```python
sns.histplot(x=df['price'], data=df)  # There may be outliers in the price column
plt.show()
```

Identifying outliers in price column

```python
sns.boxplot(y='price', data=df)   # Outliers are near 100000
plt.show()
```

```python
df['price'].mode()
```

Filtering outliers

```python
df1 = df.copy()

df1 = df1[df1['price'] <= 1500]
sns.boxplot(x='price', data=df1)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(x='price', data=df1, bins=100)
plt.ylabel('Frequency')
plt.title('Distribution of price')
plt.show()
```

Visualizing distribution for availability_365

```python
sns.histplot(x='availability_365', data=df1)
plt.title('availability_365 distribution')
plt.ylabel('Frequency')
plt.show()
```

Grouping By and analysing

```python
df1.head(1)
```

```python
# What is avg price in each neighborhood
df1['neighbourhood_group'].value_counts()
```

```python
df1.groupby(by='neighbourhood_group')['price'].mean()
```

## Feature Engineering

Price per bed analysis

```python
df1['price_per_bed'] = df1['price'] / df1['beds']
df1.head(2)
```

```python
# Grouping by neighborhood_group and avg of price_per_bed
print(df1.groupby(by='neighbourhood_group')['price_per_bed'].mean())
```

```python
print(df1.groupby(by='neighbourhood_group')['price'].mean())
```

## Bivariate Analysis

Analysing how one column affects another column

# How room type affects price in different neighborhood_group
plt.figure(figsize=(8,5))
sns.barplot(x='neighbourhood_group', y='price', data=df1, hue='room_type')
plt.show()

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='number_of_reviews', y='price', data=df1, hue='neighbourhood_group')
plt.title('Locality and review dependency on price')
plt.show()
```

Pairplot

###Geospatial Chart
